```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

```python
import math
m=[]
n=[]

for i in range(100):
    take=5+7*math.sin(i*math.pi/30)*math.cos(i*math.pi/30)
    m.append(take)
    take1=5+7*math.cos(i*math.pi/30)
    n.append(take1)

plt.axis('equal')
plt.plot(m,n)

plt.title('yayy circle')
plt.show()
```
